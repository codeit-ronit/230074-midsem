# Task 1.1 Core Contribution / Architecture

## Step-by-Step Method Description

*   **Step 1: Initialization**
    *   **Description**: The inverse temperature parameter $\beta$ is initialized to a small value $\beta_{min} \ll 1$ and the model parameter $\Theta^{(0)}$ is initialized randomly.
    *   **Reference**: Section 3.1: "1. Initialize".
    *   **Purpose**: To begin the deterministic annealing process at a high temperature where all clusters heavily overlap, preventing premature convergence to local optima.

*   **Step 2: E-step**
    *   **Description**: The posterior probabilities $h_j(t)$ are calculated using the modified scheduling parameter $\beta$, making the memberships more uniform (when $\beta < 1$) or more hard (when $\beta > 1$).
    *   **Reference**: Section 3.1: Equation (2).
    *   **Purpose**: To estimate latent memberships flexibly, allowing clusters to merge and split efficiently based on the temperature.

*   **Step 3: M-step**
    *   **Description**: The model parameters $\Theta^{(new)}$ (means, covariances, mixing coefficients) are re-estimated using the $h_j(t)$ values from the E-step, in exactly the same manner as the traditional EM algorithm.
    *   **Reference**: Section 3.1: "M-step: estimate $\Theta^{(new)}$".
    *   **Purpose**: To optimize the expected log-likelihood given the softened or hardened posterior assignments.

*   **Step 4: Temperature Updates**
    *   **Description**: The parameter $\beta$ is incrementally updated following an "anti-annealing" schedule. First it ascends from $\beta_{min}$ to 1, then continues past 1 to an upper bound $\beta_{max} > 1$, and finally descends slowly back to 1. At different $\beta$ values, the E and M steps iterate. Furthermore, noise is repeatedly injected into component means to allow distinct clusters to explicitly split.
    *   **Reference**: Section 3.1: Figure 6(b) and the paragraph discussing "Anti-annealing scheduling".
    *   **Purpose**: To push overlapping components apart by forcing harder memberships at $\beta > 1$, thereby drastically reducing component overlaps and significantly accelerating the convergence rate for unbalanced mixtures, before finally settling to optimize the true posterior at $\beta=1$.

This paper solves the problem of slow convergence of the Expectation-Maximization (EM) algorithm for Gaussian mixture models containing small, overlapping clusters (unbalanced mixing coefficients), claiming that their Deterministic Anti-Annealing schedule—which temporarily forces hardened, disjoint cluster memberships—significantly speeds up convergence and avoids poor local optima compared to standard EM or ECG optimization methods.
