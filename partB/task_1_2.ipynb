# Task 1.2 Key Assumptions

## Assumption 1
**Assumption:**
The parameter space is not completely flat during the high-temperature phases, and minor localized perturbations to the estimated means will structurally allow components to split successfully.

**Why the method needs it:**
During $\beta < 1$, the effective number of clusters is often less than the true $K$ (because components merge at high temperatures). The anti-annealing method strictly relies on adding a "small amount of random noise to the mean of each Gaussian component along the first principal component dimension" so that these merged components can actively separate and split symmetrically as $\beta$ increases.

**Violation scenario:**
A dataset scenario where the true constituent clusters share the identical first principal component but differ exclusively in higher-order (e.g. second or third) orthogonal principal components. Here, the systematic noise injection along the 1st principal component would constantly fail at properly splitting the merged cluster means toward their correct respective modes.

**Paper reference:**
Section 3.1, paragraph starting "The proposed anti-annealing schedule (Figure 6(b)) improves...", specifically: "We must perturb the estimated parameters with a small noise term to enable the clusters to split... along the first principal component dimension."

## Assumption 2
**Assumption:**
Deterministically scaling the inverse temperature $\beta > 1$ actively and reliably decreases the pairwise overlap $e_{ij}(\Theta^*)$ towards zero without causing parameter collapse.

**Why the method needs it:**
This constitutes the mathematical justification for why Anti-Annealing successfully speeds up convergence. By artificially forcing the posterior probabilities $h_i(t)$ towards 0 or 1 (the winner-take-all behavior), the pairwise component overlap functionally drops, accelerating the condition number of the Hessian matrix and yielding superlinear convergence rates.

**Violation scenario:**
A dataset where the true underlying clusters are extremely centralized and radically lack boundary distinction (e.g., highly concentric and perfectly superimposed Gaussians with equal means but vastly different variances). Forcing hard memberships ($\beta > 1$) in this geometry would systematically drive the smaller variance parameter towards zero (creating singularities and likelihood collapse) instead of smoothly accelerating component separation.

**Paper reference:**
Section 3.1, "How Does Anti-Annealing Speed up Convergence?", paragraph stating: "During the Anti-annealing phase ($\beta > 1$), the pairwise overlap values $e_{ij}(\Theta^*)$ decreases due to relatively harder membership probabilities."

***

## Assumption 3
**Assumption:**
The true data distribution consists of pure overlapping components whose geometric boundaries can be logically separated using rigid, exponentiated membership probabilities.

**Why the method needs it:**
The method fundamentally forces harder assignments between samples and clusters (via $\beta \to \infty$ approximations) to speed up parameter optimization. If clusters cannot meaningfully partition samples spatially, the artificially hardened boundaries will misguide the interim parameter updates toward suboptimal local islands.

**Violation scenario:**
A dataset that naturally follows a heavy-tailed distribution (e.g., Student-t distribution) where genuine "outliers" of one cluster heavily infiltrate the spatial core of another. Hardening the membership at $\beta > 1$ would explicitly and falsely assign these thick heavy tails entirely to the adjacent cluster core, severely corrupting the interim mean and covariance updates before the algorithm can arrive back at $\beta=1$.

**Paper reference:**
Primarily Section 1: "The Gaussian mixture model is a powerful model... It models the data as a mixture of multiple Gaussian distributions," combined with Section 3.1, Equation (2), which exponentiates normal Gaussian probability density values.
